# FINDING SOCCER-ENTHUSED COUNTRIES 

I intend on finding countries that are huge soccer fans by utilizing a technique that will compare the country's share of global world cup TV Audience with the country's share of the global population in percentage. If the country's share of global world cup TV audience is greater than or equal to country's share of global population, then I consider that a very soccer-enthused country. 

```python
pd.set_option('display.max_rows', None)
team = df
team = team.drop(["confederation","gdp_weighted_share"], axis="columns")
def get_soccerfans():
    population_share = team["population_share"]
    tv_audience_share = team["tv_audience_share"]
    
    population_share = population_share.tolist()
    tv_audience_share = tv_audience_share.tolist()
    result_list = []
    
    for i in range(len(population_share)):
        
        if tv_audience_share[i]>=population_share[i]:
            result = "fan"
            result_list.append(result)
        else:
            result = "Not a fan"
            result_list.append(result)
    
    result_column = pd.DataFrame(result_list, columns=['status'])
    soccerfanlist = team.drop(["population_share","tv_audience_share"],axis="columns")
    soccerfanlist = pd.concat([soccerfanlist,result_column],axis="columns")
    
    fanatic_countries = soccerfanlist[soccerfanlist["status"]=='fan']
    fanatic_countries = fanatic_countries.drop(['status'],axis="columns")
    fanatic_countries_list = fanatic_countries["country"].tolist()
    
    non_fanatic_countries = soccerfanlist[soccerfanlist["status"]=="Not a fan"]    
    non_fanatic_countries = non_fanatic_countries.drop(['status'],axis="columns")
    non_fanatic_countries_list = non_fanatic_countries["country"].tolist()
    
    print("These",len(fanatic_countries_list),"countries are huge soccer fans:","\n\n",fanatic_countries_list)
    print("\n\n\n\n","These" ,len(non_fanatic_countries_list)," countries do not like soccer as much:","\n\n",non_fanatic_countries_list)

get_soccerfans()
```

# RANKING THE TOP SOCCER-ENTHUSED COUNTRIES

Essentially, I will find the tv_audience_share/population_share ratios which would help me identify which countries like to watch FIFA soccer the most. 

## Countries with very low population

In order to deal with countries that have very low tv_audience share (~0), I will be substituting the tv_audience share values with a value very close to 0 (i.e. $10^{-5}$). Also, due to the imprecision of the data collected from the extremely small countries (population wise), it would be a better idea to entirely remove the countries that have a population share of 0% from my analysis as they introduce abnormal behavior.

#### AFC participants

```python
PowerDf = merged.drop(["confederation","gdp_weighted_share"], axis="columns")
afc = PowerDf[PowerDf["AFC"]==1]
afc = afc.drop(["AFC","CAF","CONCACAF","CONMEBOL", "OFC", "UEFA"], axis= "columns")
afc['population_share'] = df['population_share'].replace([0.0],0.00001)
afc['tv_audience_share'] = df['tv_audience_share'].replace([0.0],0.00001)
afc = afc[afc["population_share"]!=0.00001]
afc = afc.reset_index(drop=True)
afc_populationshare = afc.drop(["country","tv_audience_share"], axis="columns")
afc_populationshare_list = afc["population_share"].tolist()
afc_tvaudienceshare = afc.drop(["country","population_share"], axis="columns")
afc_tvaudience_list = afc["tv_audience_share"].tolist()
divided_values = []
for i in range(len(afc_tvaudience_list)):
    divided_values.append(afc_tvaudience_list[i]/afc_populationshare_list[i])

#Two dataframe columns to be concatenated
divided_values = pd.DataFrame(divided_values, columns=['ratio'])
afc = afc.drop(["population_share","tv_audience_share"], axis="columns")
afc = afc.reset_index(drop=True)
newlist = pd.concat([afc, divided_values], axis=1)
newlist = newlist.sort_values(by='ratio',ascending=False)
newlist = newlist.reset_index(drop=True)
print("The rankings of the countries in AFC are:","\n\n", newlist)
```

# COMPARING IF HIGH POPULATION MEANS MORE TV-VIEWERSHIP

Since we will be looking at all the countries that participate in FIFA, this graph/plot will not be sorted by federation. I will be plotting TV-audience versus population.

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import math

population_share = df.drop(["country","confederation","tv_audience_share","gdp_weighted_share"], axis= "columns")
population_share_list = population_share["population_share"].tolist()
population_share_list = np.array(population_share_list)
population_share_list = np.reshape(population_share_list, (191, 1))

tv_audience = df.drop(["country","confederation","population_share","gdp_weighted_share"], axis= "columns")
tv_audience_list = tv_audience["tv_audience_share"].tolist()
tv_audience_list = np.array(tv_audience_list)

linreg = LinearRegression(fit_intercept=False)
obj = linreg.fit(population_share_list, tv_audience_list)
trendline = linreg.predict(population_share_list)
plt.scatter(population_share_list, tv_audience_list)
plt.plot(population_share_list, trendline, color="black")
plt.xlabel("population percentage")
plt.ylabel("tv audience percentage")
plt.grid()
plt.show()

slope = linreg.coef_[0]

rsquaredvalue = r2_score(tv_audience_list, trendline)
r_value = math.sqrt(rsquaredvalue)
print(r_value)
```

Our r-value is 0.71. From this we can conclude that there exists a moderate level of correlation between population percentage and tv audience percentage.
